```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier()
model.fit(train_data, train_labels)
```

```python
predictions = model.predict(test_data)
```

```python
accuracy = accuracy_score(test_labels, predictions)
print("La exactitud del modelo es de:", accuracy)
```

```python
print("\nClasificación:")
print(classification_report(test_labels, predictions))
```

```python
plt.figure(figsize=(12, 8), dpi=300)
tree.plot_tree(model, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.show()
```
