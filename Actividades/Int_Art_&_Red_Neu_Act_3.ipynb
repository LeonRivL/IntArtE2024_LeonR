```python
"""Objetivo: Entrenar uno de los modelos de aprendizaje supervisado más simple: Regresión Lineal. """

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(train_data, train_labels)
```

```python
predicciones = model.predict(test_data)
```

```python
error_cuad = mean_squared_error(test_labels, predicciones)
print("El error cuadratico es:", error_cuad)
```

```python
plt.scatter(test_data, test_labels, color='red')
plt.plot(test_data, predicciones, color='black', linewidth=1)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal')
plt.show()
```
